### Primeira etapa: Data Wrangling

```python
import pandas as pd
import matplotlib.pylab as plt
```

```python
arquivo ="https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-DA0101EN-SkillsNetwork/labs/Data%20files/auto.csv"
```

```python
headers = ["symboling","normalized-losses","make","fuel-type","aspiration", "num-of-doors","body-style",
         "drive-wheels","engine-location","wheel-base", "length","width","height","curb-weight","engine-type",
         "num-of-cylinders", "engine-size","fuel-system","bore","stroke","compression-ratio","horsepower",
         "peak-rpm","city-mpg","highway-mpg","price"]
```

```python
df = pd.read_csv(arquivo, names = headers)
```

```python
df.head()
```

```python
#Substituindo "?" por NaN
```

```python
import numpy as np

df.replace("?", np.nan, inplace = True)
df.head(5)
```

```python
#Contar os dados faltantes em cada coluna:
missing_data = df.isnull()

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```

```python
#Tratamento de dados:
#Cálculo da média dos valores na coluna normalized-losses
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Média de normalized-losses:", avg_norm_loss)
```

```python
#Substituição dos valores NaN pela média calculada anteriormente
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)
```

```python
#Mesmo processo com a coluna bore:
avg_bore=df['bore'].astype('float').mean(axis=0)
print("Média de bore:", avg_bore)
```

```python
df["bore"].replace(np.nan, avg_bore, inplace=True)
```

```python
#Mesmo processo com a coluna horsepower:
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Médio horsepower:", avg_horsepower)
```

```python
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)
```

```python
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Médio peak rpm:", avg_peakrpm)
```

```python
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
avg_stroke = df["stroke"].astype("float").mean(axis = 0)
df["stroke"].replace(np.nan, avg_stroke, inplace = True)

avg_stroke = df["price"].astype("float").mean(axis = 0)
df["price"].replace(np.nan, avg_stroke, inplace = True)
```

```python
#Para contar a quantidade de valor de cada item no total das colunas:
df['num-of-doors'].value_counts()
```

```python
#Vamos substituir os valores missing pelo valor mais frequente:
df["num-of-doors"].replace(np.nan, "four", inplace=True)
```

```python
#Contar novamente os dados faltantes:
missing_data = df.isnull()

for column in missing_data.columns.values.tolist():
    print(column)
    print (missing_data[column].value_counts())
    print("")    
```
